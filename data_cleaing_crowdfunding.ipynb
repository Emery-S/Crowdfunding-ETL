```python
#get the unique values in category & sub-category
crowdfunding_info_df["category & sub-category"].unique()
```

```python
#split category & sub-category
crowdfunding_info_df[["category", "subcategory"]] = crowdfunding_info_df["category & sub-category"].str.split('/', n=1, expand=True)
crowdfunding_info_df.head(10)
```

```python
#get distinct values in category and subcategory columns.
print(crowdfunding_info_df["category"].unique())
print(crowdfunding_info_df["subcategory"].unique())
```

```python
#get unique categories and subcategories in separate lists
categories = crowdfunding_info_df["category"].unique()
subcategories = crowdfunding_info_df["subcategory"].unique()
print(categories)
print()
print(subcategories)
```

```python
#create numpy arrays with 1-10 for the category_ids and 1-25 for the subcategory_ids.
category_id = np.arange(1, 10)
subcategory_id = np.arange(1, 25)
print(category_ids)
print(subcategory_ids)
```

```python
#use a list comprehension to add "cat0" to each category_id. 
cat_ids = ["cat0" + str(cat_id) for cat_id in category_ids]

#use a list comprehension to add "scat0" to each subcategory_id. 
scat_ids = ["scat0" + str(scat_id) for scat_id in subcategory_ids ]

print(cat_ids)
print(scat_ids)
```

```python
#create a category data frame with cat_ids array as the category_id and categories list as the category name.
category_df = pd.DataFrame({'category_id': cat_ids,
                           'category': categories
                           })
subcategory_df = pd.DataFrame({'subcategory_id': scat_ids,
                              'subcategory': subcategories
                              })
subcategory_df.head(10)
```

```python
category_df.to_csv("categories.csv", index=False)
subcategory_df.to_csv("subcategories.csv", index=False)
```
